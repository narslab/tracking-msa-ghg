```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the table
data = {
    'Category': ['Mobile combustion MPY', 'Mobile combustion MPG', 'Electricity consumption',
                 'Solid waste', 'Stationary combustion', 'Agriculture',
                 'Wastewater treatment', 'Forestry'],
    'Bridgeport': [-9.77, -0.12, -25.80, 2.47, -39.41, -71.76, -20.99, -66.82],
    'Hartford': [5.37, -0.12, -45.76, -0.99, -16.92, 44.87, 81.35, -61.28],
    'New Haven': [-0.34, -0.12, -36.16, 0, -21.21, -33.08, -0.41, -67.00]
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Set the 'Category' column as the index for easier plotting
df.set_index('Category', inplace=True)

# Reverse the order of the rows
df = df.iloc[::-1]

# Plot the data with custom colors and reordered bars
colors = ['green', 'tomato', 'orange']  # Assign colors to the three cities

# Adjust the figure size (smaller than before)
fig, ax = plt.subplots(figsize=(8, 5))  # Reduced figure size

df[['Bridgeport', 'Hartford', 'New Haven']].plot(
    kind='barh',
    ax=ax,  # Use the axis from the subplots
    color=colors,  # Custom colors for each city
    width=0.8
)

# Customize the legend order
ax.legend(['Bridgeport', 'Hartford', 'New Haven'], loc='upper right', fontsize=14)

# Remove the y-axis label
ax.set_ylabel('')  # This removes the y-axis label

# Customize tick label font size
ax.tick_params(axis='x', labelsize=14)  # Font size for x-axis tick labels
ax.tick_params(axis='y', labelsize=14)  # Font size for y-axis tick labels

# Add grid lines
ax.grid(True, which='both', axis='x', color='gray', linestyle='--', linewidth=0.7)

# Adjust the margins to reduce white space (left and right)
plt.subplots_adjust(left=0.2, right=0.9)  # Reduce whitespace on the sides

ax.set_xlabel('Bias (%)', fontsize=14)  # Label for the y-axis


# Adjust the layout to avoid clipping
plt.tight_layout()
plt.savefig('../../figures/paper-images/biases.jpg', dpi=600, bbox_inches='tight')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for Bias (%) and Variance (%)
data_bias = {
    'Category': ['Mobile combustion MPY', 'Mobile combustion MPG', 'Electricity consumption',
                 'Solid waste', 'Stationary combustion', 'Agriculture',
                 'Wastewater treatment', 'Forestry'],
    'Bridgeport': [-9.77, -0.12, -25.80, 2.47, -39.41, -71.76, -20.99, -66.82],
    'Hartford': [5.37, -0.12, -45.76, -0.99, -16.92, 44.87, 81.35, -61.28],
    'New Haven': [-0.34, -0.12, -36.16, 0, -21.21, -33.08, -0.41, -67.00]
}

data_variance = {
    'Category': ['Mobile combustion MPY', 'Mobile combustion MPG', 'Electricity consumption',
                 'Solid waste', 'Stationary combustion', 'Agriculture',
                 'Wastewater treatment', 'Forestry'],
    'Bridgeport': [4.00426E-27, 1.452044926, 252.1744653, 0, 893.09813, 0, 0, 55.45971202],
    'Hartford': [4.40186E-25, 1.452044926, 438.5759664, 0, 1311.371241, 0, 0, 0.381156388],
    'New Haven': [1.24001E-24, 1.452044926, 181.8923107, 0, 599.8643332, 0, 0, 84.0643687]
}

# Create DataFrames for Bias and Variance
df_bias = pd.DataFrame(data_bias)
df_variance = pd.DataFrame(data_variance)

# Set the 'Category' column as the index for easier plotting
df_bias.set_index('Category', inplace=True)
df_variance.set_index('Category', inplace=True)

# Reverse the order of the rows for both DataFrames for plotting
df_bias = df_bias.iloc[::-1]
df_variance = df_variance.iloc[::-1]

# Define the colors for the cities
colors = ['green', 'tomato', 'orange']

# Create the subplots with shared y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Plot the Bias data
df_bias[['Bridgeport', 'Hartford', 'New Haven']].plot(
    kind='barh',
    ax=ax1,
    color=colors,
    width=0.8,
    legend=None
)
# ax1.set_title('Bias (%)', fontsize=16)
ax1.set_xlabel('Bias (%)', fontsize=14)
ax1.grid(True, which='both', axis='x', color='gray', linestyle='--', linewidth=0.7)

# Plot the Variance data
df_variance[['Bridgeport', 'Hartford', 'New Haven']].plot(
    kind='barh',
    ax=ax2,
    color=colors,
    width=0.8
)
# ax2.set_title('Variance (%)', fontsize=16)
ax2.set_xlabel('Variance (%)', fontsize=14)
ax2.grid(True, which='both', axis='x', color='gray', linestyle='--', linewidth=0.7)
ax2.legend(fontsize=14) 
# Customize the y-axis for both plots (remove y-axis label)
ax1.set_ylabel('')  # Remove y-axis label on the left side
ax2.set_ylabel('')  # Remove y-axis label on the right side

# Customize tick label font size
ax1.tick_params(axis='x', labelsize=14)
ax1.tick_params(axis='y', labelsize=14)
ax2.tick_params(axis='x', labelsize=14)
ax2.tick_params(axis='y', labelsize=14)

# Adjust the margins to reduce white space (left and right)
plt.subplots_adjust(left=0.2, right=0.9, wspace=0.1)

# Add a common legend
# handles, labels = ax1.get_legend_handles_labels()
# fig.legend(handles, labels, loc='upper right', fontsize=14)

# Adjust the layout to avoid clipping
plt.tight_layout()

# Save the facet plot
plt.savefig('../../figures/paper-images/facet_plot_bias_variance.jpg', dpi=600, bbox_inches='tight')

# Show the plot
plt.show()
```
