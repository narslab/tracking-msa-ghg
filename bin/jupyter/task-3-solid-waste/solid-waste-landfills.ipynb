```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#2021 landfill emissions data come from FLIGHT, which include landfills and resource-recoverage facility
data = np.array([['New Haven-Milford', 0],
                 ['Hartford-east Hartford-Middletown',377019/1000000],
                 ['Bridgeport-Stamford-Norwalk', 290551/1000000]])

df = pd.DataFrame(data, columns=['MSA', 'GHG Emissions (MMTCO2e)'])
df["Solid waste combustion (MMTCO2e)"]= [0,289425/1000000,290551/1000000]
df["Landfill emissions (MMTCO2e)"]= [0,87594/1000000,0]
df
df.to_csv('../../../results/solidwaste-emissions.csv')
```

```python
plt.figure(figsize=(8, 3))
X = ['Bridgeport Area','Hartford Area', 'New Haven Area']

X = X[::-1]

waste_combustion = [0, 289425/1000000, 290551/1000000]
landfills = [0, 87594/1000000, 0]

bar_width = 0.7  # Width of each bar


# Plot the stacked bars for carbon sequestration with flipped axes
plt.barh(X, waste_combustion, label='Waste combustion', height=bar_width)
plt.barh(X, landfills, left=waste_combustion, label='Landfills methane emissions', height=bar_width)


#plt.ylabel("Region", fontsize=20, labelpad=20)
plt.xlabel("Greenhouse Gas Emissions (MMTCO2e) ",fontsize=15, labelpad=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(loc='upper left', bbox_to_anchor=(0.5, 0.3),fontsize=11)

plt.savefig('../../../figures/Report-images/solid-waste-emissions.png', dpi=600, bbox_inches='tight')
plt.show()
```
