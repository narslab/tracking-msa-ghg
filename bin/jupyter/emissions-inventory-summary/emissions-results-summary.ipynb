<span style="font-family: verdana; color: green; font-size: 20px;font-weight: bold;">Step two: Import the emissions data</span>

```python
#mobile emissions
Mobile_emissions = pd.read_csv('../../../visualization data/mobile-emissions.csv')
CO2 = pd.read_csv('../../../visualization data/CO2-mobile-emissions.csv')
CO2 = CO2['CO2'].tolist()
CH4 = pd.read_csv('../../../visualization data/CH4-mobile-emissions.csv')
CH4 = CH4['CH4'].tolist()
N2O = pd.read_csv('../../../visualization data/N2O-mobile-emissions.csv')
N2O = N2O['N2O'].tolist()

Mobile_emi = Mobile_emissions
Mobile_emi.drop(Mobile_emi.columns[0], axis=1, inplace=True)

#mobile emissions breakdown by different vechile type
mobiles_breakdown = pd.read_csv('../../../visualization data/mobile-emissions-breakdown.csv')
```

```python
#electricity emissions
Electricity_emissions = pd.read_csv('../../../visualization data/electricity-emissions.csv')
Electricity_emissions.drop(Electricity_emissions.columns[0], axis=1, inplace=True)

#electricity emissions breakdown
breakdown_elec = pd.read_csv('../../../visualization data/emissions-electricity-breakdown.csv')
```

```python
#solid waste emissions
Solidwaste_emissions = pd.read_csv('../../../visualization data/solidwaste-emissions.csv')
Solidwaste_emissions.drop(Solidwaste_emissions.columns[0], axis=1, inplace=True)
Solidwaste_emissions
```

```python
#convert the emissions data into numerical value
Electricity_emissions['Residential Electricity Emissions (MMTCO2e)'] = Electricity_emissions['Residential Electricity Emissions (MMTCO2e)'].astype(float)
Electricity_emissions['Commercial/Industrial Electricity Emissions (MMTCO2e)']=Electricity_emissions['Commercial/Industrial Electricity Emissions (MMTCO2e)'].astype(float)
Electricity_emissions['Total Emissions (MMTCO2e)']=Electricity_emissions['Total Emissions (MMTCO2e)'].astype(float)
Solidwaste_emissions['Landfill methane (MMTCO2e)'] = Solidwaste_emissions['Landfill methane (MMTCO2e)'].astype(float)
Solidwaste_emissions['GHG Emissions (MMTCO2e)']=Solidwaste_emissions['GHG Emissions (MMTCO2e)'].astype(float)
Solidwaste_emissions['Solid waste combustion (MMTCO2e)']=Solidwaste_emissions['Solid waste combustion (MMTCO2e)'].astype(float)
```

```python
#residential emissions
residential_emissions = pd.read_csv('../../../visualization data/residential-energy-emissions-breakdown.csv')

#residential emissions data breakdown via different energy type
residential_breakdowm = pd.read_csv('../../../visualization data/residential-emissions-breakdown-fuel.csv')
residential_oil = residential_breakdowm.iloc[2,1:][::-1].reset_index(drop=True).astype(float)
residential_propane = residential_breakdowm.iloc[1,1:][::-1].reset_index(drop=True).astype(float)
residential_natural = residential_breakdowm.iloc[0,1:][::-1].reset_index(drop=True).astype(float)

breakdown_residential = pd.read_csv('../../../visualization data/residential-emissions-breakdown.csv')

#industrial emissions
industrial_emissions = pd.read_csv('../../../visualization data/industrial-emissions-breakdown.csv')

#commercial emissions
commercial_emissions = pd.read_csv('../../../visualization data/commercial-emissions.csv')

#agricultural emissions
AG_emissions = pd.read_csv('../../../visualization data/agriculture-emissions.csv')

#wastewater emissions
wastewater_emissions = pd.read_csv('../../../visualization data/wastewater-emissions.csv')

#urban forestry carbon sequestration
land_emissions = pd.read_csv('../../../visualization data/land-cover-emissions.csv')
```

<span style="font-family: verdana; color: green; font-size: 20px;font-weight: bold;">Step three: Concatenate emissions data </span>

```python
Total_emi = pd.Series('Total Emissions (MMTCO2e)')
TOTAL_EM = pd.concat([Total_emi,Mobile_emi["Automobile (MMTCO2e)"], Mobile_emi["Motorcycle (MMTCO2e)"], Mobile_emi["Trucks (MMTCO2e)"], Mobile_emi["Buses (MMTCO2e)"],
                       Electricity_emissions['Residential Electricity Emissions (MMTCO2e)'],Electricity_emissions['Commercial/Industrial Electricity Emissions (MMTCO2e)'],
                       Solidwaste_emissions['Solid waste combustion (MMTCO2e)'],Solidwaste_emissions['Landfill methane (MMTCO2e)'], residential_emissions['GHG emissions (MMTCO2e)'],commercial_emissions['GHG Emissions/MMTCO2e'][:3],
                     industrial_emissions['GHG emissions (MMTCO2e)'],AG_emissions['Organic fertilizer emissions (MMTCO2e)'],AG_emissions['Synthetic fertilizer emissions (MMTCO2e)'],AG_emissions['Manure fertilizer emissions (MMTCO2e)'],
                     wastewater_emissions['GHG Emissions (MMTCO2e)'],-land_emissions['Deciduous forest sequestration (MTCO2e)']/1000000,-land_emissions['Coniferous forest sequestration (MTCO2e)']/1000000,-land_emissions['Forested wetland sequestration (MTCO2e)']/1000000], axis=0)

TOTAL_EM=TOTAL_EM.tolist()
T_index = pd.MultiIndex.from_arrays(array1)
TOTAL_EM = pd.Series(TOTAL_EM, index=T_index)
```

```python
#add the CO2, CH4, and N20 emissions
co2_emi = pd.Series('CO2 Emissions (MMTCO2e)')
co2_EM = pd.concat([co2_emi,mobiles_breakdown['CO2 Emissions (MMTCO2e)'],breakdown_elec['CO2 Emissions (MMTCO2e)_re'],breakdown_elec['CO2 Emissions (MMTCO2e)_co'], pd.Series([''] *6),breakdown_residential['CO2 emissions (MMTCO2e)'],pd.Series([''] * 27)], axis=0)

#TOTAL_EM=TOTAL_EM.reset_index(drop=True)
co2_EM =co2_EM .tolist()

T_index = pd.MultiIndex.from_arrays(array1)
co2_EM  = pd.Series(co2_EM , index=T_index)


ch4_emi = pd.Series('CH4 Emissions (MMTCO2e)')
ch4_EM = pd.concat([ch4_emi,mobiles_breakdown['CH4 Emissions (MMTCO2e)'],breakdown_elec['CH4 Emissions (MMTCO2e)_re'],breakdown_elec['CH4 Emissions (MMTCO2e)_co'], pd.Series([''] *6),breakdown_residential['CH4 emissions (MMTCO2e)'],pd.Series([''] * 27)], axis=0)


#TOTAL_EM=TOTAL_EM.reset_index(drop=True)
ch4_EM =ch4_EM .tolist()

T_index = pd.MultiIndex.from_arrays(array1)
ch4_EM  = pd.Series(ch4_EM , index=T_index)

n2o_emi = pd.Series('N2O Emissions (MMTCO2e)')
n2o_EM = pd.concat([n2o_emi,mobiles_breakdown['N2O Emissions (MMTCO2e)'],breakdown_elec['N2O Emissions (MMTCO2e)_re'],breakdown_elec['N2O Emissions (MMTCO2e)_co'], pd.Series([''] *6),breakdown_residential['N2O emissions (MMTCO2e)'],pd.Series([''] * 27)], axis=0)


#TOTAL_EM=TOTAL_EM.reset_index(drop=True)
n2o_EM =n2o_EM .tolist()

T_index = pd.MultiIndex.from_arrays(array1)
n2o_EM  = pd.Series(n2o_EM , index=T_index)
```

```python
population = pd.read_csv('../../../data/tidy/population-five-towns.csv')
population = population[['Year', 'New Haven-Milford, CT Metro Area', 'Hartford-East Hartford-Middletown, CT Metro Area', 'Bridgeport-Stamford-Norwalk, CT Metro Area', 'Connecticut']]
population_20 = population.iloc[1,1:4]
population_21 = population.iloc[0,1:4]
population_17 = population.iloc[2,1:4]
population_15 = population.iloc[3,1:4]

popu = pd.Series('Population')
pop = pd.concat([popu,pd.Series(np.tile(population_21,8)), pd.Series(np.tile(population_20, 3)),pd.Series(np.tile(population_17, 3)),pd.Series(np.tile(population_21, 1)),pd.Series(np.tile(population_15, 3))], axis=0)

#TOTAL_EM=TOTAL_EM.reset_index(drop=True)
pop =pop.tolist()

T_index = pd.MultiIndex.from_arrays(array1)
pop  = pd.Series(pop, index=T_index)
```

```python
emissions_capita = TOTAL_EM[1:]*1000000/pop[1:]
#emissions_capita = TOTAL_EM*1000000/numeric_pop
```

```python
combined_series = pd.concat([pop, TOTAL_EM, emissions_capita, co2_EM, ch4_EM, n2o_EM], axis=1)
combined_series.iloc[0,2]='Emissions per capita (MTCO2e)'

# Reset the index, moving it inside the DataFrame and adding a numerical index
df_reset = combined_series.reset_index()

# Add a new numerical index
df_reset[' '] = range(1, len(df_reset) + 1)

# Set the new numerical index as the index of the DataFrame
df_reset = df_reset.set_index(' ')

# Set the first row as the column index
df_reset.columns = df_reset.iloc[0]

# Drop the first row (which is now the column index)
df_reset = df_reset.drop(df_reset.index[0])
df_reset = df_reset.reset_index(drop=True)
```
