<span style="font-family: verdana; color: green; font-size: 20px;font-weight: bold;">Step three: Import the emissions data</span>

```python
Mobile_emissions = pd.read_csv('../../../results/mobile-emissions.csv')
CO2 = pd.read_csv('../../../results/CO2-mobile-emissions.csv')
CO2 = CO2['CO2'].tolist()
CH4 = pd.read_csv('../../../results/CH4-mobile-emissions.csv')
CH4 = CH4['CH4'].tolist()
N2O = pd.read_csv('../../../results/N2O-mobile-emissions.csv')
N2O = N2O['N2O'].tolist()

Mobile_emi = Mobile_emissions
Mobile_emi.drop(Mobile_emi.columns[0], axis=1, inplace=True)

#mobile emissions breakdown by different vechile type
mobiles_breakdown = pd.read_csv('../../../results/mobile-emissions-breakdown.csv')

#electricity emissions
Electricity_emissions = pd.read_csv('../../../results/electricity-emissions.csv')
Electricity_emissions.drop(Electricity_emissions.columns[0], axis=1, inplace=True)

#electricity emissions breakdown
breakdown_elec = pd.read_csv('../../../results/emissions-electricity-breakdown.csv')

Solidwaste_emissions = pd.read_csv('../../../results/solidwaste-emissions.csv')
Solidwaste_emissions.drop(Solidwaste_emissions.columns[0], axis=1, inplace=True)

#convert the emissions data into numerical value
Electricity_emissions['Residential Electricity Emissions (MMTCO2e)'] = Electricity_emissions['Residential Electricity Emissions (MMTCO2e)'].astype(float)
Electricity_emissions['Commercial/Industrial Electricity Emissions (MMTCO2e)']=Electricity_emissions['Commercial/Industrial Electricity Emissions (MMTCO2e)'].astype(float)
Electricity_emissions['Total Emissions (MMTCO2e)']=Electricity_emissions['Total Emissions (MMTCO2e)'].astype(float)
Solidwaste_emissions['Landfill emissions (MMTCO2e)'] = Solidwaste_emissions['Landfill emissions (MMTCO2e)'].astype(float)
Solidwaste_emissions['GHG Emissions (MMTCO2e)']=Solidwaste_emissions['GHG Emissions (MMTCO2e)'].astype(float)
Solidwaste_emissions['Solid waste combustion (MMTCO2e)']=Solidwaste_emissions['Solid waste combustion (MMTCO2e)'].astype(float)

#residential emissions data breakdown via different energy type
residential_breakdowm = pd.read_csv('../../../results/residential-emissions-breakdown-fuel.csv')
residential_oil = residential_breakdowm.iloc[2,1:][::-1].reset_index(drop=True).astype(float)
residential_propane = residential_breakdowm.iloc[1,1:][::-1].reset_index(drop=True).astype(float)
residential_natural = residential_breakdowm.iloc[0,1:][::-1].reset_index(drop=True).astype(float)

#industrial emissions
industrial_emissions = pd.read_csv('../../../results/industrial-emissions-breakdown.csv')

#commercial emissions
commercial_emissions = pd.read_csv('../../../results/commercial-emissions.csv')

#agricultural emissions
AG_emissions = pd.read_csv('../../../results/agriculture-emissions.csv')

#wastewater emissions
wastewater_emissions = pd.read_csv('../../../results/wastewater-emissions.csv')

#urban forestry carbon sequestration
land_emissions = pd.read_csv('../../../results/land-cover-emissions.csv')
```

```python
#add the CO2, CH4, and N20 emissions
ghg = pd.Series('GHG Emissions (MMTCO2e)')
ghg_EM = pd.concat([ghg,Mobile_emi["Automobile (MMTCO2e)"], Mobile_emi["Motorcycle (MMTCO2e)"], Mobile_emi["Trucks (MMTCO2e)"], Mobile_emi["Buses (MMTCO2e)"],
                       Electricity_emissions['Residential Electricity Emissions (MMTCO2e)'],Electricity_emissions['Commercial/Industrial Electricity Emissions (MMTCO2e)'],
                       Solidwaste_emissions['Solid waste combustion (MMTCO2e)'],Solidwaste_emissions['Landfill emissions (MMTCO2e)'],residential_oil,residential_propane,residential_natural ,commercial_emissions['GHG Emissions/MMTCO2e'][:3],
                     industrial_emissions['GHG emissions (MMTCO2e)'],AG_emissions['Organic fertilizer emissions (MMTCO2e)'],AG_emissions['Synthetic fertilizer emissions (MMTCO2e)'],AG_emissions['Manure fertilizer emissions (MMTCO2e)'],
                     wastewater_emissions['GHG Emissions (MMTCO2e)'],-land_emissions['Deciduous forest sequestration (MTCO2e)']/1000000,-land_emissions['Coniferous forest sequestration (MTCO2e)']/1000000,-land_emissions['Forested wetland sequestration (MTCO2e)']/1000000], axis=0)

#TOTAL_EM=TOTAL_EM.reset_index(drop=True)
ghg_EM =ghg_EM.tolist()

T_index = pd.MultiIndex.from_arrays(array1)
ghg_EM  = pd.Series(ghg_EM , index=T_index)
```

```python
population = pd.read_csv('../../../results/population_five_towns.csv')
population = population[['Year', 'New Haven-Milford, CT Metro Area', 'Hartford-East Hartford-Middletown, CT Metro Area', 'Bridgeport-Stamford-Norwalk, CT Metro Area', 'Connecticut']]
population_20 = population.iloc[1,1:4]
population_21 = population.iloc[0,1:4]
population_17 = population.iloc[2,1:4]
population_15 = population.iloc[3,1:4]

popu = pd.Series('Population')
pop = pd.concat([popu,pd.Series(np.tile(population_21,8)), pd.Series(np.tile(population_20, 5)),pd.Series(np.tile(population_17, 3)),pd.Series(np.tile(population_21, 1)),pd.Series(np.tile(population_15, 3))], axis=0)

#TOTAL_EM=TOTAL_EM.reset_index(drop=True)
pop =pop.tolist()

T_index = pd.MultiIndex.from_arrays(array1)
pop  = pd.Series(pop, index=T_index)

#emissions per capita
emissions_capita = ghg_EM[1:]*1000000/pop[1:]
```

```python
combined_series = pd.concat([TOTAL_activity,ghg_EM, pop,emissions_capita], axis=1)
combined_series.iloc[0,3]='Emissions per Capita (MTCO2e)'

# Reset the index, moving it inside the DataFrame and adding a numerical index
df_reset = combined_series.reset_index()

# Add a new numerical index
df_reset[' '] = range(1, len(df_reset) + 1)

# Set the new numerical index as the index of the DataFrame
df_reset = df_reset.set_index(' ')

# Set the first row as the column index
df_reset.columns = df_reset.iloc[0]

# Drop the first row (which is now the column index)
df_reset = df_reset.drop(df_reset.index[0])
Summary = df_reset.reset_index(drop=True)
```
