```python
import pandas as pd
import numpy as np
import os
import string
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import StrMethodFormatter
%matplotlib inline
```

```python
summary = pd.read_excel('../../../visualization data/Emissions-Reduction-Calculation.xlsx', sheet_name='calculation')
```

```python
# Assuming df_new_haven_sorted is your DataFrame, sort the data. 
df_new_haven_sorted = summary.sort_values(by='Percentage of reduction %', ascending=True)

categories = df_new_haven_sorted['Symbol']
values = df_new_haven_sorted['Percentage of reduction %']

fig, ax = plt.subplots(figsize=(5, 10))

# Calculate the width of the bars based on the 'Total Emissions (MMTCO2e)' values
bar_widths = 0.8

# Create the first bar plot
ax.barh(categories, values, height=bar_widths, color='skyblue')
ax.set_xlabel('% of reduction to the total net emissions', color='black',fontsize=15)
ax.tick_params(axis='x', labelcolor='black',labelsize=10)
ax.grid(axis='x', linestyle='--', alpha=1)
ax.tick_params(axis='x', labelcolor='black',labelsize=17)

#ax1.xaxis.labelpad = 20  # You can adjust the value as needed
ax.xaxis.labelpad = 10  # You can adjust the value as needed
ax.set_xticks(np.arange(0, values.max() + 0.5, 1))
ax.tick_params(axis='y', labelcolor='black', labelsize=20)  # Adjust the labelsize as needed

# Show the plot
plt.savefig('../../../figures/new-haven-emissions-reduction.png', dpi=600, bbox_inches='tight')
```
