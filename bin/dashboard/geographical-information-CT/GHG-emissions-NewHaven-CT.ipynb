```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the pie chart
labels = ['Transportation', 'Electricity Production & Consumption', 'Energy Efficiency and Buildings', 'Industrial','Waste Management','Agricultural Working Lands']
sizes = [2.88, 1.31, 2.32, 0.34,0.09,0.03]


# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add lines pointing to each legend
for i, (label, size) in enumerate(zip(labels, sizes)):
    angle = sum(sizes[:i]) + size / 2
    x = 1.5 * np.cos(angle * 2 * np.pi / sum(sizes))
    y = 1.5 * np.sin(angle * 2 * np.pi / sum(sizes))
    plt.annotate(label, xy=(x, y), xytext=(1.35*x, 1.35*y), arrowprops=dict(facecolor='black', arrowstyle='-'))

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart with Lines Pointing to Legends')
plt.show()
```
