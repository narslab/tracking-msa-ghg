```python
# import all data manipulation libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd


# emmision data
emission_data = pd.read_csv('../../results/Activity-Emissions-Summary.csv')
Hartford_east_Hartford_Middletown  = gpd.read_file('QGIS-processed- shapefile/Hartford-east Hartford-Middletown MSA.shp')
Connecticut = gpd.read_file('QGIS-processed- shapefile/Connecticut.shp')
Hartford_City = gpd.read_file('QGIS-processed- shapefile/Hartford City.shp')
Bridgeport_Stamford_Norwalk_MSA = gpd.read_file('QGIS-processed- shapefile/Bridgeport-Stamford-Norwalk MSA.shp')
New_Haven_City = gpd.read_file('QGIS-processed- shapefile/New Haven City.shp')
# New Haven-Milford MSA
New_Haven_Milford_MSA = gpd.read_file('QGIS-processed- shapefile/New Haven-Milford MSA.shp')
# Waterbury City
Waterbury_City = gpd.read_file('QGIS-processed- shapefile/Waterbury City.shp')

# for all shapefiles, convert column name NAME10 to NAME
for shapefile in [Hartford_east_Hartford_Middletown, Connecticut, Hartford_City, Bridgeport_Stamford_Norwalk_MSA, New_Haven_City, New_Haven_Milford_MSA, Waterbury_City]:
    shapefile.rename(columns={'NAME10':'NAME'}, inplace=True)

# concatenate all the shapefiles
# this line shows all county lines in the shapefiles but we only need the ones in the next line
# counties_shapefile = pd.concat([Hartford_east_Hartford_Middletown, Connecticut, Hartford_City, Bridgeport_Stamford_Norwalk_MSA, New_Haven_City, New_Haven_Milford_MSA, Waterbury_City], ignore_index=True)

counties_shapefile = pd.concat([Hartford_east_Hartford_Middletown, Connecticut, Bridgeport_Stamford_Norwalk_MSA, New_Haven_Milford_MSA], ignore_index=True)
# rename column "New Haven-Milford" to "New Haven", "Hartford-West Hartford-East Hartford" to "Hartford, Middlesex, Tolland", "Bridgeport-Stamford-Norwalk" to "Fairfield"
counties_shapefile['NAME'] = counties_shapefile['NAME'].str.replace('New Haven-Milford', 'New Haven')
counties_shapefile['NAME'] = counties_shapefile['NAME'].str.replace('Hartford-West Hartford-East Hartford', 'Hartford, Middlesex, Tolland')
counties_shapefile['NAME'] = counties_shapefile['NAME'].str.replace('Bridgeport-Stamford-Norwalk', 'Fairfield')


# remove ", CT" from the NAME column
counties_shapefile['NAME'] = counties_shapefile['NAME'].str.replace(', CT', '')

# convert counties_shapefile to csv and save in same directory
counties_shapefile.to_csv('county_shapefile.csv')

# add a row to the emission data that contains the County ENTRY "Connecticut"
new_row = pd.DataFrame({'County':'Connecticut'}, index=[0])

# concatenate the new row to the emission data
emission_data = pd.concat([new_row, emission_data], ignore_index=True)


display(emission_data.head(10))
display(emission_data.County.unique())
print(emission_data.shape)

display(counties_shapefile.head(10))
print(counties_shapefile.shape)
```

```python
import pandas as pd
import json
from shapely.geometry import shape, Polygon, MultiPolygon
from shapely.ops import unary_union

# Assuming emission_data and counties_shapefile are already defined and loaded
data_with_counties = pd.merge(emission_data, counties_shapefile, left_on='County', right_on='NAME')

# Convert geometry to GeoJSON
data_with_counties['geometry_geojson'] = data_with_counties['geometry'].apply(lambda x: x.__geo_interface__)


# Simplify MultiPolygons to Polygons by selecting the largest polygon (this is to convert MultiPolygons to Polygons which is the only type of geometry that can be visualized on Superset)
def simplify_multipolygon_to_polygon(geojson):
    geom = shape(geojson)
    if isinstance(geom, MultiPolygon):
        # Convert the MultiPolygon to a list of polygons and select the largest by area
        largest_polygon = max(geom.geoms, key=lambda x: x.area)
        simplified_geojson = largest_polygon.__geo_interface__
    else:
        simplified_geojson = geojson  # No change if not a MultiPolygon
    return simplified_geojson


data_with_counties['simplified_geometry_geojson'] = data_with_counties['geometry_geojson'].apply(simplify_multipolygon_to_polygon)

# Transform the simplified geometry to a complete GeoJSON object
def wrap_geojson(geometry):
    # Ensure the geometry is wrapped as a GeoJSON Feature
    geojson_full = json.dumps({
        "type": "Feature",
        "geometry": geometry,
        "properties": {}  # Empty properties, modify as needed
    })
    return geojson_full

data_with_counties['complete_geojson'] = data_with_counties['simplified_geometry_geojson'].apply(wrap_geojson)

# save to csv
data_with_counties.to_csv('emission_data_with_counties.csv')

# Verify the transformation
data_with_counties[['geometry_geojson', 'complete_geojson']].head()
```

```python
import pydeck as pdk
import pandas as pd

# Sample data: Replace this with your DataFrame
# Assuming your DataFrame 'df' has a column 'geometry_geojson_str' with GeoJSON strings for geometries
data = {
    'name': ['Area 1', 'Area 2'],
    'geometry_geojson_str': [
        '{"type": "Polygon", "coordinates": [[[30, 10], [40, 40], [20, 40], [10, 20], [30, 10]]]}',
        '{"type": "Polygon", "coordinates": [[[15, 5], [25, 35], [5, 35], [-5, 15], [15, 5]]]}'
    ]
}
df = pd.DataFrame(data)

# Convert the GeoJSON string column to a format that PyDeck can interpret
df['geometry'] = df['geometry_geojson_str'].apply(lambda x: eval(x))

# Define a PyDeck layer
# Here, we use the PolygonLayer to plot polygons, but you can use other layers like ScatterplotLayer for points
layer = pdk.Layer(
    'PolygonLayer',  # Change the layer type as needed
    df,
    get_polygon='geometry.coordinates',  # Path to the coordinates
    get_fill_color=[180, 0, 200, 140],  # RGBA color of the polygons
    get_line_color=[0, 0, 0, 200],  # RGBA color of the polygon borders
    get_line_width=2,  # Width of the polygon borders
    pickable=True,
    auto_highlight=True,
)

# Set the viewport location
view_state = pdk.ViewState(latitude=20, longitude=25, zoom=3, bearing=0, pitch=0)

# Render the deck.gl map
r = pdk.Deck(layers=[layer], initial_view_state=view_state, map_style='mapbox://styles/mapbox/light-v9')

# Display the map
r.show()
```
